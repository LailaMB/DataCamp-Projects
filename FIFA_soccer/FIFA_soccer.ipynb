```python
import pandas as pd
fifa = pd.read_csv("data/fifa.csv") 
fifa.head()
```

```python
fifa.describe()
```

```python
fifa.info()
```

```python
#remove spaces in columns names
fifa.columns = fifa.columns.str.strip()

#remove spaces in position values
fifa['position'] = fifa['position'].str.strip()

# The possible positions are: 'GK' (goalkeeper), 'M' (midfield), 'A' (attack) and 'D' (defense). 
fifa['position'].unique()
```

```python
# Heights of the goalkeepers: gk_heights
gk_heights= fifa['height'][fifa['position']== 'GK']

# Heights of the other players: other_heights
other_heights=fifa['height'][fifa['position'] != 'GK']

# Print out the mean height of goalkeepers.
print("Mean height of goalkeepers: " + str(gk_heights.mean()))

# Print out the mean height of other players. Replace 'None'
print("Mean height of other players: " + str(other_heights.mean()))
```
