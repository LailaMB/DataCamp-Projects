```python
import pandas as pd

weather = pd.read_csv('pittsburgh2013.csv', index_col='Date', parse_dates=True)
weather.head()
```

```python
weather.loc['2013-7-1':'2013-7-7', 'PrecipitationIn']
```

```python
week1_range = weather.loc['2013-07-01':'2013-07-07', ['Min TemperatureF', 'Max TemperatureF']]
print(week1_range)
```

```python
week1_mean = weather.loc['2013-07-01':'2013-07-07','Mean TemperatureF']
print(week1_mean)
```

```python
## Relative temprature range 
week1_range.divide(week1_mean, axis='rows')
```

```python
# Percentage changes

# subtracting the previuose day's value from the current day's value and dividing by the previouse day value.
week1_mean.pct_change() * 100
```

#### Convert degrees Fahrenheit to degrees Celsius

```python
# Write a function to convert degrees Fahrenheit to degrees Celsius: to_celsius
def to_celsius(F):
    return 5/9*(F - 32)

# Apply the function over 'Mean TemperatureF' and 'Mean Dew PointF': df_celsius
df_celsius = weather[['Mean TemperatureF','Mean Dew PointF']].apply(to_celsius)

# Reassign the columns df_celsius
df_celsius.columns = ['Mean TemperatureC', 'Mean Dew PointC']

# Print the output of df_celsius.head()
df_celsius.head()
```

```python
# Extract selected columns from weather as new DataFrame: temps_f
temps_f = weather[['Min TemperatureF', 'Mean TemperatureF','Max TemperatureF']]

# Convert temps_f to celsius: temps_c
temps_c = (temps_f - 32) * 5/9

# Rename 'F' in column names with 'C': temps_c.columns
temps_c.columns = temps_c.columns.str.replace('F', 'C')

# Print first 5 rows of temps_c
print(temps_c.head())
```

```python
# Print the mean of the January and March data
january = weather.loc['2013-01','Mean TemperatureF']
march = weather.loc['2013-03','Mean TemperatureF']


print(january.mean(), march.mean())

# Print the standard deviation of the January and March data
print(january.std(), march.std())

```
